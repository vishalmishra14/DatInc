```python
df_Crime_County = df_Crime[['lat','long','state','killed', 'injured', 'casualties']]
```

```python
plt.figure(figsize = (24,12))

m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution='l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')

xpt, ypt = m(lon, lat)
m.plot(xpt, ypt, 'r*', markersize=15)
```

```python
plt.title("Crimes in states")
plt.show()
```
